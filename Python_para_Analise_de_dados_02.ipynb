# **Python para Analise de Dados - Estatísticas**

```python
import pandas as pd
```

```python
arquivo = '/content/kc_house_data.csv'
dataset = pd.read_csv(arquivo,sep=',',header=0)
```

**Estatistica Descritivas**

```python
# Valor médio da coluna bedrooms
dataset['bedrooms'].mean()
```

```python
# Valor máximo da coluna bedrooms
dataset['bedrooms'].max()
```

```python
# Valor mínimo da coluna bedrooms
dataset['bedrooms'].min()
```

```python
# desvio padrão da coluna bedrooms
dataset['bedrooms'].std()
```

**sobre simetria** : 
* um valor zero indica uma distribuição simétrica 
* um valor maior que zero ou menor indica uma distribuição assimétrica
* valores acima de zero podemos dizer que existe uma assimetria positiva
* valores abaixo de zero uma assimetria negativa <br>

Logo, valores muita acima de zero indicam que existem mais valores acima da média e valores abaixo de zero significa que contém mais valores abaixo da média











```python
# Simetria da coluna bedrooms
dataset['bedrooms'].skew()
```

```python
# Retorna simetria de cada coluna do dataset
dataset.skew()
```

**Correlação**

```python
# Correlação de todas as colunas (técnica person)
# Se uma variavel esta correlacionada positivamente, se uma váriavel aumenta a outra tende a aumentar também
# Se uma variavel esta correlacionada negativamente, se uma aumenta a outra tende a diminuir
dataset.corr()
```

```python
# correlação de spearman (leva em consideração a positiva e a negativa)
dataset.corr('spearman')
```

```python
# Correlação de algumas colunas
dataset[['bedrooms','bathrooms','floors','waterfront','grade','price']].corr()
```

**Tabelas pivot**


```python
# Tabelas Pivot são úteis para fazer agrupamento
# Aqui realizamos uma contagem de imóveis agrupados pelas colunas waterfront e floors
dataset.pivot_table('id',index=['waterfront','floors'],aggfunc='count',margins=True)
```

**Crosstab**

```python
# Cruza valores das variáveis
# Aqui podemos ver a distribuição dos imóveis por número de quartos com relação a sua condição
pd.crosstab(dataset['bedrooms'],dataset['condition'])
```
